```python
def build_patient_dict(data_dir):
    
    json_data = open(data_dir)
    amd_data = json.load(json_data)
    patient_dict = dict()
    
    count_idx = 1
    for idx, record in enumerate(amd_data):
        if idx % 100 == 0:
            print("{} patients processed...".format(count_idx*100))
            count_idx += 1
            
        this_pid = record["ID2"]
        this_visits = record["VISITS"]
        this_clinical_feature = []
        this_clinical_feature.append(record["ENROLLAGE"])
        this_clinical_feature.append(record["SEX"])
        this_clinical_feature.append(record["BMI_R"])
        
        visitno = []
        feature_re = []
        feature_le = []
        severe_score_re = []
        severe_score_le = []
        scale = []
        
        for visit in this_visits:
            this_visitno = int(visit["VISNO"])
            this_feature_le = []
            this_feature_re = []
            this_scale = []
            
            if len(visit) >= 24:
                visitno.append(this_visitno)
                # append features
                this_feature_re.append(visit["REDRSZWI"])
                this_feature_le.append(visit["LEDRSZWI"])
                this_feature_re.append(visit["REGEOAWI"])
                this_feature_le.append(visit["LEGEOAWI"])
                this_feature_re.append(visit["REGEOACT"])
                this_feature_le.append(visit["LEGEOACT"])
                this_feature_re.append(visit["REGEOACS"])
                this_feature_le.append(visit["LEGEOACS"])
                this_feature_re.append(visit["RESUBFF2"])
                this_feature_le.append(visit["LESUBFF2"])
                this_feature_re.append(visit["RENDRUF2"])
                this_feature_le.append(visit["LENDRUF2"])
                this_feature_re.append(visit["RESSRF2"])
                this_feature_le.append(visit["LESSRF2"])
                this_feature_re.append(visit["RESUBHF2"])
                this_feature_le.append(visit["LESUBHF2"])
                this_feature_re.append(visit["RERPEDWI"])
                this_feature_le.append(visit["LERPEDWI"])
                this_feature_re.append(visit["REINCPWI"])
                this_feature_le.append(visit["LEINCPWI"])
                
                feature_re.append(this_feature_re)
                feature_le.append(this_feature_le)
                severe_score_re.append(visit["AMDSEVRE"])
                severe_score_le.append(visit["AMDSEVLE"])
                scale.append(visit["SCALE"])
        
        patient_dict[this_pid] = {"visitno" : visitno, "clinical_feature" : this_clinical_feature, "feature_re" : feature_re, 
                                  "feature_le" : feature_le, "severe_score_re" : severe_score_re, "severe_score_le" : severe_score_le,
                                 "scale" : scale}
    
    return patient_dict
```

```python
def remove_nan(mylist):
    
    myarray = np.array(mylist)
    removed_nan_idx = ~np.isnan(myarray)
    
    return myarray[removed_nan_idx]

def count_decreasing_scale(amd_patient_dict):
    
    decreasing_patient_id = []
    decreasing_patient_num = 0
    
    for key, value in amd_patient_dict.items():
        scale = remove_nan(value["scale"])
        
        is_decreasing = False
        
        for i in range(len(scale)-1):
            if scale[i+1] >= scale[i]:
                continue
            else:
                is_decreasing = True
                
        if is_decreasing:
            decreasing_patient_id.append(key)
            decreasing_patient_num += 1
    
    print("{} patient have decreasing scale".format(decreasing_patient_num))
    
    return decreasing_patient_id
```

```python
def remove_8(feature_list, return_idx=False):
    
    feature_list_removed_8 = []
    remaining_idx = []
    
    for idx, feature in enumerate(feature_list):
        
        this_feature = np.array(feature)
        value_8_ind = np.where(this_feature == 8)
        count_value_8 = np.sum(value_8_ind)
        
        if count_value_8 == 0:
            feature_list_removed_8.append(feature)
            remaining_idx.append(idx)
        else:
            continue
            
    if return_idx:
        return feature_list_removed_8, remaining_idx
    else:
        return feature_list_removed_8
    
def remove_recurring_label(label_list, target_label):
    """
    remove the recurring value except the first value from the list
    """
    
    label_array = np.array(label_list)
    count_target_label = len(np.where(label_array == target_label)[0])

    if count_target_label > 0:
        
        target_label_idx = np.where(label_array == target_label)[0]
        target_label_first_idx = target_label_idx[0]
        return label_array[:target_label_first_idx+1], target_label_first_idx+1
    
    else:
        return label_array, len(label_array)
    
def convert_severe_score(severe_score_list, cutoff_score, label_type):
    
    if label_type == "binary":
        severe_score_array = np.array(severe_score_list)
        binary_severe_score = np.where(severe_score_array < cutoff_score, 0, 1)
        return binary_severe_score
    
    else:
        severe_score_array = np.array(severe_score_list)
        clipped_severe_score = np.where(severe_score_array <= cutoff_score, severe_score_array, cutoff_score)
        return clipped_severe_score
        
def build_eye_dict(patient_dict, cutoff_score, label_type, filter_single_eye):
    """
    label type must be either binary or score
    """
    
    if label_type not in ["binary", "score"]:
        raise ValueError("label_type must be either binary or score")
    
    amd_eye_dict = dict()
    
    for key, value in patient_dict.items():
        
        amd_eye_dict[key] = dict()
        
        this_clinical_feature = value["clinical_feature"]
        this_feature_re = value["feature_re"]
        this_feature_le = value["feature_le"]
        
        this_feature_re, remaining_idx_re = remove_8(this_feature_re, return_idx=True)
        this_feature_le, remaining_idx_le = remove_8(this_feature_le, return_idx=True)
        
        this_severe_score_re = np.array(value["severe_score_re"])
        this_severe_score_le = np.array(value["severe_score_le"])
        this_severe_score_re = this_severe_score_re[remaining_idx_re]
        this_severe_score_le = this_severe_score_le[remaining_idx_le]
        
        if label_type == "binary":
            this_severe_score_re = convert_severe_score(this_severe_score_re, cutoff_score, label_type="binary")
            this_severe_score_le = convert_severe_score(this_severe_score_le, cutoff_score, label_type="binary")
            
            this_severe_score_re, remaining_idx_re = remove_recurring_label(this_severe_score_re, 1.)
            this_severe_score_le, remaining_idx_le = remove_recurring_label(this_severe_score_le, 1.)
            this_feature_re = this_feature_re[:remaining_idx_re]
            this_feature_le = this_feature_le[:remaining_idx_le]
            
        if label_type == "score":
            this_severe_score_re = convert_severe_score(this_severe_score_re, cutoff_score, label_type="score")
            this_severe_score_le = convert_severe_score(this_severe_score_le, cutoff_score, label_type="score")
            
            this_severe_score_re, remaining_idx_re = remove_recurring_label(this_severe_score_re, cutoff_score)
            this_severe_score_le, remaining_idx_le = remove_recurring_label(this_severe_score_le, cutoff_score)
            this_feature_re = this_feature_re[:remaining_idx_re]
            this_feature_le = this_feature_le[:remaining_idx_le]
            
        this_re_dict = dict()
        this_le_dict = dict()
        
        this_re_dict["clinical_feature"] = this_clinical_feature
        this_re_dict["feature"] = this_feature_re
        this_re_dict["severe_score"] = this_severe_score_re
        
        this_le_dict["clinical_feature"] = this_clinical_feature
        this_le_dict["feature"] = this_feature_le
        this_le_dict["severe_score"] = this_severe_score_le
        
        if len(this_re_dict["severe_score"]) > 1:
            amd_eye_dict[key]["re"] = this_re_dict
        if len(this_le_dict["severe_score"]) > 1:
            amd_eye_dict[key]["le"] = this_le_dict
            
    if filter_single_eye:
    
        count_two_eyes = 0
        count_one_eye = 0
        amd_two_eyes_dict = dict()
    
        for key, value in amd_eye_dict.items():
        
            if len(value) == 2:
                count_two_eyes += 1
                amd_two_eyes_dict[key] = value
            else:
                count_one_eye += 1
    
        print("{} patients have two eyes data".format(count_two_eyes))
        print("{} patients have one eye data".format(count_one_eye))
        return amd_two_eyes_dict
    
    return amd_eye_dict

def convert_clinical_feature(eye_dict):
    
    age = []
    bmi = []
    converted_eye_dict = dict()
    
    for key, value in eye_dict.items():
        age.append(value["re"]["clinical_feature"][0])
        bmi.append(value["re"]["clinical_feature"][2])
    
    max_age, min_age, mean_age = np.nanmax(age), np.nanmin(age), np.nanmean(age)
    max_bmi, min_bmi, mean_bmi = np.nanmax(bmi), np.nanmin(bmi), np.nanmean(bmi)
    
    for key, value in eye_dict.items():
        
        converted_eye_dict[key] = value
        
        converted_clinical_feature = []
        this_age = value["re"]["clinical_feature"][0]
        this_sex = value["re"]["clinical_feature"][1]
        this_bmi = value["re"]["clinical_feature"][2]
        
        if np.isnan(this_age):
            this_minimaxed_age = minmax_normalization(mean_age, max_age, min_age)
        else:
            this_minimaxed_age = minmax_normalization(this_age, max_age, min_age)
        converted_clinical_feature.append(this_minimaxed_age)
            
        if this_sex == "M":
            this_encoded_sex = [1, 0]
        else:
            this_encoded_sex = [0, 1]
        converted_clinical_feature.extend(this_encoded_sex)
            
        if np.isnan(this_bmi):
            this_minimaxed_bmi = minmax_normalization(mean_bmi, max_bmi, min_bmi)
        else:
            this_minimaxed_bmi = minmax_normalization(this_bmi, max_bmi, min_bmi)
        converted_clinical_feature.append(this_minimaxed_bmi)
        
        converted_eye_dict[key]["re"]["clinical_feature"] = converted_clinical_feature
        converted_eye_dict[key]["le"]["clinical_feature"] = converted_clinical_feature
    
    return converted_eye_dict
    
def impute_nan(mylist):
    myarray = np.array(mylist)
    nanmean = np.nanmean(myarray)
    
    return np.where(~np.isnan(myarray), myarray, nanmean) 

def minmax_normalization(value, max_value, min_value):
    return (value - min_value) / (max_value - min_value)
        

def save_data(output_path, mydata):
    with open(output_path, 'wb') as f:
        pickle.dump(mydata, f)
```

```python
def count_late_amd(eye_dict, late_amd_label):
    
    late_amd_count = 0
    
    for key, value in eye_dict.items():
        
        is_this_patient_late_amd = False
        
        for k, v in value.items():
            this_severe_score = v["severe_score"]
            late_amd_ind = np.where(this_severe_score == late_amd_label)[0]
            
            if len(late_amd_ind) > 0:
                is_this_patient_late_amd = True
        
        if is_this_patient_late_amd:
            late_amd_count += 1
    
    print("{} patients developed late AMD".format(late_amd_count))
```

```python
def impute_nan(mylist):
    myarray = np.array(mylist)
    nanmean = np.nanmean(myarray)
    
    return np.where(~np.isnan(myarray), myarray, nanmean) 

def minmax_normalization(mylist):
    myarray = np.array(mylist)
    max_value = np.max(myarray)
    min_value = np.min(myarray)
    
    return (myarray - min_value) / (max_value - min_value)

def build_data_dict(amd_eye_dict):
    
    age = []
    sex = []
    bmi = []
    clinical_feature = []
    cfp_feature = []
    severe_score = []
    
    for key, value in amd_eye_dict.items():
        age.append(value["clinical_feature"][0])
        if value["clinical_feature"][1] == "M":
            sex.append([1.,0.]) # Male is converted to [1, 0]
        else:
            sex.append([0.,1.]) # Female is converted to [0, 1]
        bmi.append(value["clinical_feature"][2])
        
        cfp_feature.append(value["feature"])   
        severe_score.append(value["severe_score"])
    
    age = impute_nan(age)
    bmi = impute_nan(bmi)
    normalized_age = minmax_normalization(age)
    normalized_bmi = minmax_normalization(bmi)
    
    for i in range(len(age)):
        this_clinical_feature = []
        this_clinical_feature.append(normalized_age[i])
        this_clinical_feature.append(normalized_bmi[i])
        this_clinical_feature.extend(sex[i])
        clinical_feature.append(this_clinical_feature)
        
    return [clinical_feature, cfp_feature, severe_score]

def save_data(output_path, mydata):
    with open(output_path, 'wb') as f:
        pickle.dump(mydata, f)
```

```python
def label_patients(patient_dict, visit_window, target_scale):
    
    label_1_count = 0
    label_0_count = 0
    
    for key, value in patient_dict.items():
        visitno_list = np.array(value["visitno"])
        scale_list = np.array(value["scale"])

        target_visit_idx = np.array(visitno_list >= visit_window[0]) * np.array(visitno_list <= visit_window[1])
        scale_in_window = scale_list[target_visit_idx]
        scale_test = np.sum(scale_in_window == target_scale)
        
        if scale_test == 0:
            this_label = 0
            label_0_count += 1
        else:
            this_label = 1
            label_1_count += 1
            
        patient_dict[key]["label"] = this_label
    
    print("0 label : {}".format(label_0_count))
    print("1 label : {}".format(label_1_count))
    
    return patient_dict
```

```python
def replace_value(mylist, value):
    myarray = np.array(mylist)
    myarray_value_tf = (myarray != value) * 1
    
    return myarray * myarray_value_tf

def impute_nan(mylist):
    myarray = np.array(mylist)
    nanmean = np.nanmean(myarray)
    
    return np.where(~np.isnan(myarray), myarray, nanmean) 

def minmax_normalization(mylist):
    myarray = np.array(mylist)
    max_value = np.max(myarray)
    min_value = np.min(myarray)
    
    return (myarray - min_value) / (max_value - min_value)

def get_baseline_statistics(target_pateint_dict):
    
    age_list = []
    bmi_list = []
    male_count = 0
    female_count = 0
    
    for key, value in target_pateint_dict.items():
        age_list.append(value["clinical_feature"][0])
        if value["clinical_feature"][1] == "M":
            male_count += 1
        else:
            female_count += 1
        bmi_list.append(value["clinical_feature"][2])
        
    
    print("median age: {}".format(np.nanmedian(age_list)))
    print("75 percentile age: {}".format(np.nanpercentile(age_list, 75)))
    print("25 percentile age: {}".format(np.nanpercentile(age_list, 25)))
    print("median bmi: {}".format(np.nanmedian(bmi_list)))
    print("75 percentile bmi: {}".format(np.nanpercentile(bmi_list, 75)))
    print("25 percentile bmi: {}".format(np.nanpercentile(bmi_list, 25)))
    print("{} male".format(male_count))
    print("{} female".format(female_count))

def data_pipeline(target_pateint_dict, visit_limit):
    
    age_list = []
    sex_list = []
    bmi_list = []
    clinical_feature_list = []
    cfp_feature_list = []
    label_list = []
    
    for key, value in target_pateint_dict.items():
        age_list.append(value["clinical_feature"][0])
        if value["clinical_feature"][1] == "M":
            sex_list.append([1.,0.])
        else:
            sex_list.append([0.,1.])
        bmi_list.append(value["clinical_feature"][2])
        
        visit_in_window_idx = np.array(value["visitno"]) <= visit_limit
        
        visit_in_window = []
        for i in range(np.sum(visit_in_window_idx)):
            this_visit = value["feature"][i]
            this_visit = replace_value(this_visit, 8)
            visit_in_window.append(list(this_visit))
            
        cfp_feature_list.append(visit_in_window)   
        label_list.append(value["label"])
    
    age_list = impute_nan(age_list)
    bmi_list = impute_nan(bmi_list)
    age_list = minmax_normalization(age_list)
    bmi_list = minmax_normalization(bmi_list)
    
    for i in range(len(age_list)):
        this_clinical_feature = []
        this_clinical_feature.append(age_list[i])
        this_clinical_feature.append(bmi_list[i])
        this_clinical_feature.extend(sex_list[i])
        clinical_feature_list.append(this_clinical_feature)
        
    return [clinical_feature_list, cfp_feature_list, label_list]
```
