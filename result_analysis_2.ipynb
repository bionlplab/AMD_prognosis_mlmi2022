```python
def get_max_validation_auc(result_dict):
    
    max_validation_auc = []
    
    for key, mydict in result_dict.items():
        max_validation_auc.append(np.max(mydict["validation_AUC"]))
        
    return max_validation_auc
```

```python
def get_entire_auc(result_dict):
    
    max_entire_auc = []
    
    for key, mydict in result_dict.items():
        max_entire_auc.append(mydict["result"]["entire_AUC"])
        
    return max_entire_auc
```

```python
def calculate_AUC_confidence_interval(repetition_result_dict, ci=0.95):
    
    auc_list = []
    
    for key, value in repetition_result_dict.items():
        auc_list.append(value["result"]["entire_AUC"])
        
    confidence_interval = st.t.interval(ci, len(auc_list)-1, loc=np.mean(auc_list), scale=st.sem(auc_list))
    
    return np.mean(auc_list), confidence_interval
```

```python
def calculate_AUC_confidence_interval_resnet_feature_evaluation(repetition_result_dict, ci=0.95):
    
    auc_list = []
    
    for key, value in repetition_result_dict.items():
        auc_list.append(value["entire_AUC"])
        
    confidence_interval = st.t.interval(ci, len(auc_list)-1, loc=np.mean(auc_list), scale=st.sem(auc_list))
    
    return np.mean(auc_list), confidence_interval
```

```python
def calculate_perlength_AUC_confidence_interval(repetition_result_dict, ci=0.95, unique_length=5):
    
    auc_dict = dict()
    auc_ci_dict = dict()
    
    for length in range(unique_length):
        auc_dict[length+1] = []
        auc_ci_dict[length+1] = []
    
    for key, value in repetition_result_dict.items():
        auc_dict[1].append(value["result"][1]["AUC"])
        auc_dict[2].append(value["result"][2]["AUC"])
        auc_dict[3].append(value["result"][3]["AUC"])
        auc_dict[4].append(value["result"][4]["AUC"])
        auc_dict[5].append(value["result"][5]["AUC"])
    
    for key, value in auc_dict.items():
        confidence_interval = st.t.interval(ci, len(value)-1, loc=np.mean(value), scale=st.sem(value))
        auc_ci_dict[key].append((np.mean(value), confidence_interval))
    
    return auc_ci_dict
```

```python
def display_label_dist(data_list):
    
    elem_label_list = []
    
    for elem in data_list:
        if np.sum(elem) > 0:
            elem_label_list.append(1)
        else:
            elem_label_list.append(0)
        
    return np.unique(elem_label_list, return_counts=True)
```

```python
def summarize_five_fold_result(fold_path, file_name):
    fold_list = os.listdir(fold_path)
    
    entire_auc_list = []
    per_length_1_auc_list = []
    per_length_2_auc_list = []
    per_length_3_auc_list = []
    per_length_4_auc_list = []
    per_length_5_auc_list = []
    
    for fold in fold_list:
        file_path = os.path.join(fold_path, fold, file_name)
        result_dict = load_data(file_path)
        
        entire_auc_list.append(result_dict["entire_AUC"])
        per_length_1_auc_list.append(result_dict[1]["AUC"])
        per_length_2_auc_list.append(result_dict[2]["AUC"])
        per_length_3_auc_list.append(result_dict[3]["AUC"])
        per_length_4_auc_list.append(result_dict[4]["AUC"])
        per_length_5_auc_list.append(result_dict[5]["AUC"])
        
    print("mean entire AUC: {}".format(np.mean(entire_auc_list)))
    print("std entire AUC: {}".format(np.std(entire_auc_list)))
    print("-----------------------------------------")
    print("mean AUC: {}".format(np.mean(per_length_1_auc_list)))
    print("std AUC: {}".format(np.std(per_length_1_auc_list)))
    print("-----------------------------------------")
    print("mean AUC: {}".format(np.mean(per_length_2_auc_list)))
    print("std AUC: {}".format(np.std(per_length_2_auc_list)))
    print("-----------------------------------------")
    print("mean AUC: {}".format(np.mean(per_length_3_auc_list)))
    print("std AUC: {}".format(np.std(per_length_3_auc_list)))
    print("-----------------------------------------")
    print("mean AUC: {}".format(np.mean(per_length_4_auc_list)))
    print("std AUC: {}".format(np.std(per_length_4_auc_list)))
    print("-----------------------------------------") 
    print("mean AUC: {}".format(np.mean(per_length_5_auc_list)))
    print("std AUC: {}".format(np.std(per_length_5_auc_list)))
```
